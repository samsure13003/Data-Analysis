### Feature Engineering

#### Date and Time

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

##### Changing the data type of Date,Month and Year because they are still object

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
## Drop Date of Journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(3)
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_Minute'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Minute'] = df['Departure_Minute'].astype(int)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Extract hours
df['Duration_Hour'] = (
    df['Duration']
    .str.extract(r'(\d+)h')
    .fillna(0)
    .astype(int)
)

# Extract minutes
df['Duration_Minute'] = (
    df['Duration']
    .str.extract(r'(\d+)m')
    .fillna(0)
    .astype(int)
)

# df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
# df['Duration_Minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.drop('Duration', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna('1 stop')

df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
})
```

#### dropping Route column because we have 2 other columns Source and Destination

```python
df.drop('Route', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
df['Additional_Info'] = df['Additional_Info'].str.strip().str.lower()
```

```python
encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray()
```

```python
encoded_df = pd.DataFrame(
    encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray(),
    columns=encoder.get_feature_names_out(['Airline', 'Source', 'Destination', 'Additional_Info'])
)

# Merge encoded features back into the main dataframe
# Keep the original columns if you want, or drop them first

df = pd.concat([df.drop(columns=['Airline', 'Source', 'Destination', 'Additional_Info']), encoded_df], axis=1)
```
